# Quantum Dynamics from Driving Fields

```python
# Importing necessary packages
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.linalg import expm
```

## Rabi Oscillations, Two-Level Atom

```python
X = np.array([[0, 1], [1, 0]])
N = np.array([[0, 0], [0, 1]])

twopi = 2 * np.pi

def H_func(Delta, Omega):
    return - (Delta * N + Omega * X)

def evolve_state(psi_0, H, t):
    return np.matmul(expm(-1j * H * t), psi_0)
```

## Resonance: $\Delta = 0$

```python
Delta, Omega = 0 * twopi, 0.5 * twopi
H = H_func(Delta, Omega)
psi_0 = np.array([[1], [0]])
times = np.linspace(0, 5, 100)
states = [evolve_state(psi_0, H, t) for t in times]
plt.plot(times, [np.absolute(psi[0])**2 for psi in states])
plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$P(|0\rangle)$', fontsize=16)
plt.show()
```

## Far detuning: $\Delta \gg \Omega$

```python
Delta, Omega = 0.4 * twopi, 0.04 * twopi
H = H_func(Delta, Omega)
psi_0 = np.array([[1], [0]])
times = np.linspace(0, 5, 100)

states = [evolve_state(psi_0, H, t) for t in times]
plt.plot(times, [np.absolute(psi[0])**2 for psi in states])
plt.ylim((0.7, 1.03))
plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$P(|0\rangle)$', fontsize=16)
plt.show()
```

## $\Delta \approx \Omega$

```python
Delta, Omega = 0.4 * twopi, 0.3 * twopi
H = H_func(Delta, Omega)
psi_0 = np.array([[1], [0]])
times = np.linspace(0, 5, 100)

states = [evolve_state(psi_0, H, t) for t in times]
plt.plot(times, [np.absolute(psi[0])**2 for psi in states])
plt.ylim((0, 1.1))
plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$P(|0\rangle)$', fontsize=16)
plt.show()
```

## Two Two-Level Atoms

```python
import qutip as qt
from qutip.operators import *
```

```python
I = identity(2)
N = qt.Qobj(np.array([[0, 0], [0, 1]]))

def solve_me(Delta, Omega, V, times):
    # Define the Hamiltonian
    H = - Delta * (qt.tensor(N, I) + qt.tensor(I, N)) + \
        - Omega / 2 * (qt.tensor(sigmax(), I) + qt.tensor(I, sigmax())) + \
        + V * qt.tensor(N, N)
    
    # Collapse operators
    c_ops = []
    
    # Operators to measure
    m_ops = []
    
    # Initial state
    psi_0 = qt.tensor(qt.basis(2, 0), qt.basis(2, 0))
    
    # Solve master equation
    res = qt.mesolve(H, psi_0, times, c_ops , m_ops)
    return res.states
```

## Resonance, No Interaction: $\Delta = V = 0$

```python
# Parameters
Delta, Omega, V = 0, 0.5 * twopi, 0

# Time to plot for
times = np.linspace(0, 5, 100)
result = solve_me(Delta, Omega, V, times)

# Reduced density matrix of qubit A for all time points
rho_1 = [result[i].ptrace(0) for i in range(len(result))]

probs = [1]

# Calculate the bloch vector of qubit A for all time points
bloch_vec = qt.expect(sigmaz(), rho_1)**2

# Plot the result
plt.plot(times, bloch_vec)
plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$P(|0\rangle)$', fontsize=16)
plt.show()
```

## Resonance, Interaction

```python
# Parameters
Delta, Omega, V = 0, 0.3 * twopi, 2

# Time to plot for
times = np.linspace(0, 20, 200)
result = solve_me(Delta, Omega, V, times)

# Reduced density matrix of qubit A for all time points
rho_1 = [result[i].ptrace(0) for i in range(len(result))]

# Calculate the bloch vector of qubit A for all time points
bloch_vec = qt.expect(sigmaz(), rho_1)**2

# Plot the result
plt.plot(times, bloch_vec)
plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$P(|0\rangle)$', fontsize=16)
plt.show()
```
