```python
import rebound
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
```

```python
sim = rebound.Simulation()
sim.G = 4 * np.pi ** 2
sim.units = ['AU', 'yr', 'Msun']
```

```python
sim.add('Sun')
```

```python
sim.particles[0].m
```

```python
sim.add('Mercury')
sim.add('Venus')
sim.add('Earth')
```

```python
sim.add("'Oumuamua")
```

```python
fig2 = rebound.OrbitPlot(sim, unitlabel='[Au]', plotparticles=[1, 2, 3], xlim=[-1.2, 1.2], ylim=[-1.2, 1.2])
```

```python
fig = rebound.OrbitPlot(sim, unitlabel='[Au]', xlim=[-5,25], ylim=[-2.5,10])
```

```python
sun = sim.particles[0]
mercury = sim.particles[1]
venus = sim.particles[2]
earth = sim.particles[3]
oumuamua = sim.particles[4]
```

```python
oumuamua.inc
```

```python
Noutputs = 200
times = np.linspace(0, -5, Noutputs)
x = np.zeros((2, Noutputs))
y = np.zeros((2, Noutputs))
z = np.zeros((2, Noutputs))
for i, time in enumerate(times):
    sim.integrate(time)
    x[0][i] = mercury.x
    y[0][i] = mercury.y
    z[0][i] = mercury.z
    x[1][i] = oumuamua.x
    y[1][i] = oumuamua.y
    z[1][i] = oumuamua.z
```

```python
distance = np.sqrt(np.square(x[0] - x[1]) + np.square(y[0] - y[1]))
plt.plot(times, distance)
plt.xlabel('Time[yrs]', fontsize=18)
plt.ylabel('Distance [AU]', fontsize=18)
```

```python
import datetime
closeencounterdistance = np.argmin(distance)
closeencountertime = times[closeencounterdistance]
encounterdate = datetime.datetime.today() + datetime.timedelta(days=365.25 * closeencountertime)
print(distance[closeencounterdistance])
encounterdate.strftime('%Y-%m-%d %H:%M')
```
