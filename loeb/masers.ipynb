# Constraining the Spin of Black Holes Based on Lense-Thirring Precession of Maser Disks

```python
# Importing libraries
import matplotlib
matplotlib.rcParams['mathtext.default'] = 'rm'

import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from astropy import units as u
from astropy import constants as cst

# Function to calculate Chi Max
def chi_max(r, m, t):
    return cst.c**3 * r**3 / (2 * cst.G**2 * m**2 * t)

def chi_max2(r, m, t):
    return cst.c**3 * r**3 / (6 * cst.G**2 * m**2 * t)
```

## Data Points from MCP IX

```python
galaxies = [
    'NGC 4258',
    'NGC 4388',
    'NGC 2273',
    'IC 2560',
    'UGC 3789',
    'NGC 1194',
    'NGC 3393',
    'J0437+2456',
    'NGC 2960',
    'NGC 5495',
    'NGC 6323',
    'ESO 558-G009',
    'NGC 5765b',
    'NGC 6264',
    'UGC 6093'
]

masses = np.array([4.00, 0.84, 0.75, 0.44, 1.04, 6.5, 3.10, 0.29, 1.16, 1.05, 0.94, 1.70, 4.55, 2.91, 2.65]) * 10**7;
    
radii = np.array([0.11, 0.24, 0.03, 0.11, 0.08, 0.54, 0.17, 0.06, 0.13, 0.10, 0.13, 0.20, 0.30, 0.24, 0.12]);

outer_radii = np.array([0.29, 0.29, 0.084, 0.47, 0.30, 1.33, 0.60, 0.13, 0.37, 0.30, 0.30, 0.47, 1.15, 0.80, 0.24]);
```

```python
galaxies
```

```python
T = (1 * 10**10 * u.yr).to(u.s)
chi_max_values = np.zeros((1, len(galaxies))).tolist()[0]
chi_max_values2 = np.zeros((1, len(galaxies))).tolist()[0]
for i, galaxy in enumerate(galaxies):
    chi_max_values[i] = chi_max((radii[i] * u.pc).to(u.m), (masses[i] * u.M_sun).to(u.kg), T)
    chi_max_values2[i] = chi_max2((radii[i] * u.pc).to(u.m), (masses[i] * u.M_sun).to(u.kg), T) 
    
# Plotting the values
plt.figure(figsize=(8, 8))
plt.scatter(masses, chi_max_values)
plt.ylim((0, 1))
plt.xlim((0, 6 * 10**7))
plt.xlabel(r'SMBH Mass [$10^7   \mathrm{M_\odot}$]', fontsize=24)
plt.ylabel(r'$\chi_{max}$', fontsize=24)
plt.title(r'Maximum Spin (T = $1 \times 10^{10} yr)$', fontsize=24)
for chi in chi_max_values:
    print(chi, end=', ')
    
plt.figure(figsize=(8, 8))
plt.scatter(masses, chi_max_values2)
plt.ylim((0, 1))
plt.xlim((0, 6 * 10**7))
plt.xlabel(r'SMBH Mass [$10^7   \mathrm{M_\odot}$]', fontsize=24)
plt.ylabel(r'$\chi_{max}$', fontsize=24)
plt.title(r'Maximum Spin (T = $1 \times 10^{10} yr)$', fontsize=24)
for chi in chi_max_values2:
    print(chi, end=', ')
```

```python
T = (1 * 10**10 * u.yr).to(u.s)
chi_max_values_outer = np.zeros((1, len(galaxies))).tolist()[0]
for i, galaxy in enumerate(galaxies):
    chi_max_values_outer[i] = chi_max((outer_radii[i] * u.pc).to(u.m), (masses[i] * u.M_sun).to(u.kg), T)
    
# Plotting the values
plt.figure(figsize=(8, 8))
plt.scatter(masses, chi_max_values_outer)
plt.ylim((0, 1))
plt.xlim((0, 8 * 10**7))
plt.xlabel(r'SMBH Mass [$10^7   \mathrm{M_\odot}$]', fontsize=24)
plt.ylabel(r'$\chi_{max}$', fontsize=24)
plt.title(r'Maximum Spin (T = $1 \times 10^{10} yr)$', fontsize=24)
for chi in chi_max_values_outer:
    print(chi, end=', ')
```

```python
# Figure with annotations
plt.figure(figsize=(8, 8))
plt.scatter(masses, chi_max_values)
plt.ylim((0, 1))
plt.xlim((0, 6 * 10**7))
plt.xlabel(r'SMBH Mass [$10^7   \mathrm{M_\odot}$]', fontsize=24)
plt.ylabel(r'$|\chi_{max}|\frac{T}{10^{10 yr}}$', fontsize=24)

for i, galaxy in enumerate(galaxies):
    plt.annotate(f'   {galaxy}', (masses[i], chi_max_values[i]))
```
